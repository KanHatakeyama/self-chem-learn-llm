```python
import pandas as pd
```

```python

parsed_records=[]
for csv_path in ["BradleyMeltingPointDataset_clean.csv",]:
    df=pd.read_csv(csv_path)
    records=df.to_dict(orient="records")

    for compound_data in records:
        if "name" not in compound_data:
            compound_data["name"]="unknown"
        n_dict={
            "CompName":compound_data["name"],
            "SMILES":compound_data["smiles"],
            "Property":"Melting temperature",
            "Value":compound_data["mpC"],
            "unit":"[oC]", 
        }

        n_dict["Source"]="BradleyMeltingPointDataset"
        parsed_records.append(n_dict)
```
